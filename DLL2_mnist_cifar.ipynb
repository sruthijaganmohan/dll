# **MNIST dataset**

```python
# imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras import models, layers, datasets
import tensorflow as tf
```

```python
# load dataset
(X1_train, y1_train), (X1_test, y1_test) = datasets.mnist.load_data()
```

```python
print("X_train shape: ", X1_train.shape)
print("y_train shape: ", y1_train.shape)
print("X_test shape: ", X1_test.shape)
print("y_test shape: ", y1_test.shape)
```

```python
# displaying one image (pixel values range from 0 to 255)
pd.DataFrame(X1_train[12])
```

```python
# normalization
X1_train = X1_train/255
X1_test = X1_test/255
```

```python
# define model
model1 = models.Sequential()
model1.add(layers.Flatten(input_shape=(28, 28, 1)))
model1.add(layers.Dense(32, activation="relu"))
model1.add(layers.Dense(16, activation="relu"))
model1.add(layers.Dense(10, activation="softmax"))
```

```python
# compile model
model1.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
```

```python
# model summary
model1.summary()
```

```python
# train model
history1 = model1.fit(X1_train, y1_train, epochs=12, validation_data=(X1_test, y1_test))
```

```python
# plot accuracy
plt.plot(history1.history['accuracy'], label='accuracy')
plt.plot(history1.history['val_accuracy'], label='val_accuracy')
plt.title("Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
```

```python
# plot loss
plt.plot(history1.history['loss'], label='loss')
plt.plot(history1.history['val_loss'], label='val_loss')
plt.title("Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
```

# **CIFAR 10**

```python
# load dataset
cifar10 = tf.keras.datasets.cifar10
(X2_train, y2_train), (X2_test, y2_test) = cifar10.load_data()
```

```python
# normalization
X2_train = X2_train/255
X2_test = X2_test/255
```

```python
# define model
model2 = models.Sequential()
model2.add(layers.Flatten(input_shape=(32, 32, 3)))
model2.add(layers.Dense(32, activation="relu"))
model2.add(layers.Dense(16, activation="relu"))
model2.add(layers.Dense(10, activation="softmax"))
```

```python
# compile model
model2.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
```

```python
model2.summary()
```

```python
# train model
history2 = model2.fit(X2_train, y2_train, epochs=12, validation_data=(X2_test, y2_test))
```

```python
# plot accuracy
plt.plot(history2.history['accuracy'], label='accuracy')
plt.plot(history2.history['val_accuracy'], label='val_accuracy')
plt.title("Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
```

```python
# plot loss
plt.plot(history2.history['loss'], label='loss')
plt.plot(history2.history['val_loss'], label='val_loss')
plt.title("Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
```
